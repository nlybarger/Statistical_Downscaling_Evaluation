```python


import pickle
import re
import numpy as np
import funcs_DS_eval as fdse


# Cluster specific modules

from itertools import combinations

import funcs_weather_typing as fwt


def varlev(str):
    var = re.findall(r'[a-zA-Z]+', str)
    levstr = re.findall(r'[0-9]+', str)
    if levstr:
        lev = float(levstr[0])*100.
    else:
        lev = None
    return var[0],lev


def rmse(predictions, targets, nan_policy = 'raise'):
    if nan_policy == 'omit':
        return np.nanmean(np.sqrt(((predictions - targets) ** 2)))
```

```python
regions, _= fdse.make_buko_region_map('1deg')
VariableList = ['U850', 'U500', 'U250', 
            'V850', 'V500', 'V250', 
            'UV850', 'UV500', 'UV250', 
            'Q850', 'Q500', 'ZG500', 'MFL500',
            'PW', 'PSL']
combinations_list = []
for r in range(2, 6):
    combinations_list.extend(list(comb) for comb in combinations(VariableList, r))
ncombos = len(combinations_list)
diri = '/glade/derecho/scratch/nlybarger/WT_centroids_era5/varcomp_outputs/'

test_vari = {}
zopt = {}

for reg in regions:
    regname = reg.replace(' ', '')
    fili = f'{diri}{regname}.0{ncombos-1}.VARI_varcomp.pkl'
    with open(fili, 'rb') as file:
        test_vari[reg] = pickle.load(file)

    zopt[reg] = np.argmax(test_vari[reg])
    print(reg)
    print(zopt[reg])
    print(combinations_list[zopt[reg]])
    print('===================================================')
```
